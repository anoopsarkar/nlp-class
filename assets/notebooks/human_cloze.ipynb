```python
import nltk
import random
#word_choice = 'violin'
word_choice = 'butter'
#word_choice = 'sauce'
#word_choice = 'sandwich'
#word_choice = 'probabilities'
#word_choice = 'joke'
#word_choice = 'clothing'
```

```python
def replace_with_blanks(sent, word):
    return map(lambda x: '_______' if x == word else x, sent)

def mask_random(sent):
    rand_word = random.choice(sent)
    return replace_with_blanks(sent, rand_word)

for sent in nltk.corpus.brown.sents():
    if word_choice in sent:
        masked_sents = []
        masked_sents.append(' '.join(mask_random(random.choice(nltk.corpus.brown.sents()))))
        masked_sents.append(' '.join(replace_with_blanks(sent, word_choice)))
        masked_sents.append(' '.join(mask_random(random.choice(nltk.corpus.brown.sents()))))
        random.shuffle(masked_sents)
        print('-' * 72)
        print("\n\n".join(masked_sents))
```
